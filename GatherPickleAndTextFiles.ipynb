### Imports

```python
import os
import shutil
```

### List all important Directories

```python
libri_dir = './LibriSpeech100/'
train_dir = libri_dir + 'train/train-clean-100/'
dev_dir = libri_dir + 'dev/dev-clean/'
test_dir = libri_dir + 'test/test-clean/'
```

### Destination Folders

```python
train_dest = libri_dir + 'train/train_all/'
dev_dest = libri_dir + 'dev/dev_all/'
test_dest = libri_dir + 'test/test_all/'
```

### Create Lists of all Folders

```python
train_list = sorted(os.listdir(train_dir))
dev_list = sorted(os.listdir(dev_dir))
test_list = sorted(os.listdir(test_dir))
```

### Gather all Training Files
Copies all .pkl and .txt files to the final destination

```python
for folder in train_list:
    print(folder)
    folder_list = os.listdir(train_dir+folder)
    for folder2 in folder_list:
        print(folder+'/'+folder2)
        file_list = os.listdir(train_dir+folder+'/'+folder2)
        for file in file_list:
            if file.endswith(".txt") or file.endswith(".pkl"):
                src = train_dir+folder+'/'+folder2+'/' + file
                shutil.copy(src, train_dest)
```

### Gather all Dev Files

```python
for folder in dev_list:
    print(folder)
    folder_list = os.listdir(dev_dir+folder)
    for folder2 in folder_list:
        print(folder+'/'+folder2)
        file_list = os.listdir(dev_dir+folder+'/'+folder2)
        for file in file_list:
            if file.endswith(".txt") or file.endswith(".pkl"):
                src = dev_dir+folder+'/'+folder2+'/' + file
                shutil.copy(src, dev_dest)
```

### Gather all Test Files

```python
for folder in test_list:
    print(folder)
    folder_list = os.listdir(test_dir+folder)
    for folder2 in folder_list:
        print(folder+'/'+folder2)
        file_list = os.listdir(test_dir+folder+'/'+folder2)
        for file in file_list:
            if file.endswith(".txt") or file.endswith(".pkl"):
                src = test_dir+folder+'/'+folder2+'/' + file
                shutil.copy(src, test_dest)
```
